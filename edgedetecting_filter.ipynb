```python
import cv2
import scipy
import numpy as np
from scipy import misc
img = misc.ascent()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
x_shape=img.shape[0]
print('x_shape=',x_shape)
y_shape=img.shape[1]
print('y_shape=',y_shape)

img_transformed=np.zeros((x_shape,y_shape))
plt.imshow(img_transformed)
```

```python
filter = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1]]
```

```python
for i in range (1,x_shape-1):
    for j in range(1,y_shape-1):
        convolution=0.0
        convolution=convolution+(img[i-1][j-1])*(filter[0][0])
        convolution=convolution+(img[i][j-1])*(filter[0][1])
        convolution=convolution+(img[i+1][j-1])*(filter[0][2])
        convolution=convolution+(img[i-1][j])*(filter[1][0])
        convolution=convolution+(img[i][j])*(filter[1][1])
        convolution=convolution+(img[i+1][j])*(filter[1][2])
        convolution=convolution+(img[i-1][j+1])*(filter[2][0])
        convolution=convolution+(img[i][j+1])*(filter[2][1])
        convolution=convolution+(img[i+1][j+1])*(filter[2][2])
        if(convolution < 0):
            convolution = 0
        if( convolution > 255):
            convolution = 255
        img_transformed[i][j]=convolution

        
```

```python
plt.gray()
plt.grid(False)
plt.imshow(img_transformed)
```
